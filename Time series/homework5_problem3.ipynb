```python
import pandas as pd

# Load mystery data from the CSV file
data = pd.read_csv('mystery_data.csv')

# Extract time series A and B
A = data['A'].values
B = data['B'].values
```

```python
import matplotlib.pyplot as plt

# Plot time series A and B
plt.figure(figsize=(14, 7))

plt.subplot(2, 1, 1)
plt.plot(A, label='A')
plt.title('Time Series A')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(B, label='B')
plt.title('Time Series B')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
## 3a: Plot the ACF and PACF for time series A and B. Choose a model class for A and a model class for B. Justify your choice. (enter answer below)
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

fig, axes = plt.subplots(2, 2, figsize=(14, 7))

plot_acf(A, ax=axes[0, 0], title='ACF - Series A')
plot_pacf(A, ax=axes[0, 1], title='PACF - Series A')
plot_acf(B, ax=axes[1, 0], title='ACF - Series B')
plot_pacf(B, ax=axes[1, 1], title='PACF - Series B')

plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
## 3b: Fit your proposed models for time series A and B respectively and obtain model estimates.

# AR(3) for series A: order = (p, d, q) = (3, 0, 0)
model_A = ARIMA(A, order=(3, 0, 0)).fit()
print("AR(3) Model for Series A")
print(model_A.summary())

print("\n" + "="*80 + "\n")

# MA(2) for series B: order = (p, d, q) = (0, 0, 2)
model_B = ARIMA(B, order=(0, 0, 2)).fit()
print("MA(2) Model for Series B")
print(model_B.summary())
```

```python
# 3c) Plot the residual errors for estimating model A, and plot the ACF and the power spectrum of the residual errors.
import numpy as np

resid_A = model_A.resid

fig, axes = plt.subplots(2, 2, figsize=(14, 8))

axes[0, 0].plot(resid_A)
axes[0, 0].set_title('Residuals - AR(3) Model A')

plot_acf(resid_A, ax=axes[0, 1], title='ACF of Residuals A')

plot_pacf(resid_A, ax=axes[1, 0], title='PACF of Residuals A')

freqs_A = np.fft.rfftfreq(len(resid_A))
spectrum_A = np.abs(np.fft.rfft(resid_A))**2
axes[1, 1].plot(freqs_A, spectrum_A)
axes[1, 1].set_title('Power Spectrum of Residuals A')
axes[1, 1].set_xlabel('Frequency')

plt.tight_layout()
plt.show()
```
